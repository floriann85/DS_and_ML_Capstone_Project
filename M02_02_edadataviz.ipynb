```python
import pandas as pd

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
df = pd.read_csv(URL)
df.head(5)
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

Next, let's drill down to each site visualize its detailed launch records.


### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
sns.catplot(
    x="FlightNumber", 
    y="LaunchSite", 
    hue="Class", 
    data=df, 
    aspect=5, 
    height=5
)
plt.xlabel("Flight Number", fontsize=15)
plt.ylabel("Launch Site", fontsize=15)
plt.title("Flight Number vs. Launch Site (farblich nach Erfolg)", fontsize=16)
plt.show()
```

Now try to explain the patterns you found in the Flight Number vs. Launch Site scatter point plots.


Interpretation of the graphic:
“Flight Number vs. Launch Site” with color coding according to success (class)

Observed patterns:
1. CCAFS SLC 40 (Cape Canaveral Space Launch Complex 40)
•    Earlier flights (flight number < 20): many failures (blue dots).
•    Later flights: clear improvement – mostly successful (orange dots).
•	Pattern: Learning curve visible – more experience and higher success rate over time.

2. KSC LC 39A (Kennedy Space Center, Launch Complex 39A)
•    Only flights with higher flight numbers (from approx. 40).
•    Almost all landings successful → orange dots dominate.
•	Interpretation: Later use of this pad, apparently with more mature technology and higher success rates. 

3. VAFB SLC 4E (Vandenberg Air Force Base)
•    Fewer flights overall.
•    Mix of successes and failures.
•    No clear pattern as with the other sites – possibly less frequently used or different mission profiles.

Conclusion:

•    Experience increases success: With increasing FlightNumber (i.e., more launches), the success rate rises noticeably—especially at CCAFS SLC 40.

•	Some launch sites were only used later (e.g., KSC LC 39A) – and immediately showed higher success rates → presumably after technological improvements.

•    Launch site is a relevant influencing factor: Different success distributions depending on the launch 

### TASK 2: Visualize the relationship between Payload Mass and Launch Site


We also want to observe if there is any relationship between launch sites and their payload mass.


```python
# Plot a scatter point chart with x axis to be Pay Load Mass (kg) and y axis to be the launch site, and hue to be the class value
sns.scatterplot(
    x="PayloadMass", 
    y="LaunchSite", 
    hue="Class", 
    data=df,
    # Optional: point size 
    s=100  # Punktgröße (optional)
)

plt.xlabel("Payload Mass (kg)", fontsize=15)
plt.ylabel("Launch Site", fontsize=15)
plt.title("Payload Mass vs. Launch Site (farblich nach Erfolg)", fontsize=16)
plt.show()
```

Now if you observe Payload Mass Vs. Launch Site scatter point chart you will find for the VAFB-SLC  launchsite there are no  rockets  launched for  heavypayload mass(greater than 10000).


Interpretation of TASK 2: Payload Mass vs. Launch Site

Observation:
For VAFB-SLC 4E (Vandenberg Air Force Base), there are no rocket launches with a payload mass > 10,000 kg.

Significance:
1. Limited payload capacity or mission profile:
VAFB is often used for special missions such as polar or sun-synchronous orbits (SSO), where payloads are typically smaller.
These types of missions do not require satellites or payloads as large as those needed for GTO or LEO missions, for example.

2. Technical or logistical limitations:
The infrastructure (e.g., launch pad, transport logistics) may not be designed for very heavy payloads.

3. Historical use:

Most heavy payloads have been launched from Florida (CCAFS or KSC) over the years, as these locations are closer to the equator and therefore offer an energy advantage (more momentum due to Earth's rotation).

Conclusion:

VAFB SLC 4E has not been used for heavy payload launches (>10,000 kg) – either due to deliberate mission planning or technical/geographical circumstances.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
```

```python
# Gruppieren nach Orbit und mittlere Erfolgsrate berechnen
orbit_success = df.groupby('Orbit')['Class'].mean().reset_index()

# Barplot erstellen
sns.barplot(x='Orbit', y='Class', data=orbit_success)
plt.xticks(rotation=45)
plt.ylabel('Erfolgsrate', fontsize=12)
plt.xlabel('Orbit-Typ', fontsize=12)
plt.title('Erfolgsrate pro Orbit-Typ', fontsize=14)
plt.ylim(0, 1)
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


Analysis: Success rates per orbit type

Based on the bar chart from df.groupby(‘Orbit’)[‘Class’].mean():

Orbits with the highest success rate (close to 1.0 / 100%)

Typically (depending on the data), you will see:

•	SSO (Sun-Synchronous Orbit)

•	LEO (Low Earth Orbit)

•	PO (Polar Orbit)

These orbits often have near-perfect success rates because:

•	They usually carry smaller payloads.

•	Landings can be made over land or nearby sea areas.

•	More frequent repeat flights with routine mission profiles take place.

Orbits with lower success rates:

•	GTO (Geostationary Transfer Orbit)
→ Usually heavier payloads, more complex flight profiles, higher energy requirements

•	ISS (International Space Station)
→ May be lower, depending on the mission period (there were more frequent failures at the beginning of SpaceX flights)

Conclusion:

•	LEO/SSO missions are more reliable.

•	GTO and similar orbits require more technology and energy → higher risk.

•	The orbit type is an important predictor of the landing success of the first stage.

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
sns.scatterplot(
    x="FlightNumber", 
    y="Orbit", 
    hue="Class", 
    data=df,
    # Optional: point size
    s=100  
)

plt.xlabel("Flight Number", fontsize=12)
plt.ylabel("Orbit Type", fontsize=12)
plt.title("Flight Number vs. Orbit Type (farblich nach Erfolg)", fontsize=14)
plt.show()
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


Analysis of TASK 4: Flight Number vs. Orbit Type
LEO (Low Earth Orbit):

•	You can see that lower flight numbers (i.e., early missions) result in more failures (Class = 0).

•	As the flight number increases, the number of successful landings (Class = 1) increases significantly.

Interpretation:

Experience and repeated LEO missions lead to better results.
SpaceX has clearly learned and optimized in this orbit category.

GTO (Geostationary Transfer Orbit):

•	No clear pattern discernible.

•	Both low and high flight numbers show a mixture of failures and successes.

Interpretation:

•	Success in GTO missions does not appear to be directly dependent on the flight number.

•	GTO missions are often more complex, with heavier payloads and more specialized requirements—the learning curve effect is less visible.

Conclusion:

•	LEO: Success increases with experience → routine flights, improved systems.

•	GTO: Complexity overshadows the learning effect → success fluctuates independently of flight number.

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate


# 1. Extract the year (if not already done)
year = []
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

df['Year'] = Extract_year()

# 2. Calculate success rate per year
yearly_success = df.groupby('Year')['Class'].mean().reset_index()

# 3. Draw a line graph
plt.figure(figsize=(10,5))
sns.lineplot(x='Year', y='Class', data=yearly_success, marker='o')

plt.xlabel('Jahr', fontsize=12)
plt.ylabel('Durchschnittliche Erfolgsrate', fontsize=12)
plt.title('Jährlicher Trend der Start-Erfolgsraten', fontsize=14)
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# HINT: Use get_dummies() function on the categorical columns
# List of categorical columns to be coded
categorical_columns = ['Orbit', 'LaunchSite', 'LandingPad', 'Serial']

# Perform one-hot encoding
features_one_hot = pd.get_dummies(df, columns=categorical_columns)

# Show result
features_one_hot.head()
```

### TASK  8: Cast all numeric columns to `float64`


Now that our <code>features_one_hot</code> dataframe only contains numbers, cast the entire dataframe to variable type <code>float64</code>


```python
# HINT: use astype function
# Step 1: Identify non-numeric columns
non_numeric_cols = features_one_hot.select_dtypes(include=['object']).columns

# Step 2: Remove these columns (if they are not needed)
features_one_hot = features_one_hot.drop(non_numeric_cols, axis=1)

# Step 3: Cast to float64
features_one_hot = features_one_hot.astype('float64')

# Control
features_one_hot.dtypes.head()
```
